```python
import os
import pandas as pd
```

```python
data_path = '../data/dataset/'
data_folders = os.listdir(data_path)
file_names = [os.listdir(os.path.join(data_path, folder)) for folder in data_folders]
file_names
```

```python
files_dict = dict(zip(data_folders, file_names))
base_gcs_path = 'gs://idyllic-slice-361403.appspot.com/food_dataset'
data_array = []

for (dict_key, files_list) in files_dict.items():
    for filename in files_list:
        if '.jpg' not in filename: continue # don't include non-photos
        label = dict_key
        data_array.append((f'{base_gcs_path}/{dict_key}/{filename}', label))
```

```python
data_array
```

```python
dataframe = pd.DataFrame(data_array)
```

```python
dataframe.to_csv('food_data.csv', index=False, header=False)
```
